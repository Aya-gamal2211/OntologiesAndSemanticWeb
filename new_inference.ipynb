```python
from rdflib import Graph, Namespace
from owlrl import DeductiveClosure, OWLRL_Semantics

# Load RDF data from the ontology file
g = Graph()
g.parse("C:\\Users\\Shaima\\Desktop\\OntologiesAndSemanticWeb\\owl_new.ttl", format="ttl")

# Apply OWL reasoning to infer implicit triples
DeductiveClosure(OWLRL_Semantics).expand(g)

# Define the namespace
ns = Namespace("http://www.semanticweb.org/shaima/ontologies/2024/3/untitled-ontology-2#")

def check_cardinality_constraint(property_uri, min_cardinality):
    query = f"""
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    PREFIX ns: <http://www.semanticweb.org/shaima/ontologies/2024/3/untitled-ontology-2#>
    ASK WHERE {{
        ?subject rdf:type ns:Movie .
        {{
            SELECT (COUNT(?object) AS ?count) WHERE {{
                ?subject ns:{property_uri} ?object .
            }}
        }}
        FILTER (?count < {min_cardinality})
    }}
    """
    result = g.query(query)
    return not bool(result)

# Check if Movie instances have at least one actor
actor_constraint_satisfied = check_cardinality_constraint("hasActor", 1)
print("Actor cardinality constraint satisfied:", actor_constraint_satisfied)

# Check if Movie instances have at least two genres
genre_constraint_satisfied = check_cardinality_constraint("hasGenre", 2)
print("Genre cardinality constraint satisfied:", genre_constraint_satisfied)

# Check if Movie instances have at least two writers
writer_constraint_satisfied = check_cardinality_constraint("hasWriter", 2)
print("Writer cardinality constraint satisfied:", writer_constraint_satisfied)
```
